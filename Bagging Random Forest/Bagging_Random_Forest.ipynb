```python
from pandas import read_csv
from sklearn.model_selection import KFold
from sklearn.model_selection import cross_val_score
from sklearn.ensemble import BaggingClassifier
from sklearn.tree import DecisionTreeClassifier
```

```python
# Load the dataset
filename = '/content/sample_data/pima-indians-diabetes.csv'
class_names = ['preg', 'plas', 'pres', 'skin', 'test', 'mass', 'pedi', 'age', 'class']
dataframe = read_csv(filename, names=class_names)
dataframe
```

```python
array = dataframe.values
```

```python
array
```

```python
X = array[:, 0:8]
Y = array[:, 8]
```

```python
X
```

```python
Y
```

```python
# Set up the parameters
seed = 7
kfold = KFold(n_splits=10, shuffle=True, random_state=seed)  # Shuffle enabled
cart = DecisionTreeClassifier()
num_trees = 100
```

```python
# Bagging model
model = BaggingClassifier(estimator=cart, n_estimators=num_trees, random_state=seed)
results = cross_val_score(model, X, Y, cv=kfold)
print(results.mean())
```

```python
from pandas import read_csv
from sklearn.model_selection import KFold
from sklearn.model_selection import cross_val_score
from sklearn.ensemble import RandomForestClassifier

# Assume the dataset is already loaded in `array`
X = array[:, 0:8]
Y = array[:, 8]

# Random Forest parameters
num_trees = 100
max_features = 3

# KFold with shuffle enabled
kfold = KFold(n_splits=10, shuffle=True, random_state=7)

# Random Forest Model
model = RandomForestClassifier(n_estimators=num_trees, max_features=max_features)

# Evaluate model using cross-validation
results = cross_val_score(model, X, Y, cv=kfold)

# Print the mean accuracy
print(results.mean())
```
